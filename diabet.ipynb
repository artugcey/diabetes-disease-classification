```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
data.value_counts()
```

```python
preg = data.iloc[:,0:1].values
glu = data.iloc[:,1:2].values
blood = data.iloc[:,2:3].values
skin = data.iloc[:,3:4].values
ins = data.iloc[:,4:5].values
BMI = data.iloc[:,5:6].values
Diab = data.iloc[:,6:7].values
Age = data.iloc[:,7:8].values
out = data.iloc[:,8:9].values
```

```python
Preg = pd.DataFrame(data=preg,index=range(768),columns=['Preg'])
Glu = pd.DataFrame(data=glu,index=range(768),columns=['Glu'])
Blood = pd.DataFrame(data=blood,index=range(768),columns=['Blood'])
Skin = pd.DataFrame(data=skin,index=range(768),columns=['Skin'])
İns = pd.DataFrame(data=ins,index=range(768),columns=['İns'])
Bmı = pd.DataFrame(data=BMI,index=range(768),columns=['Bmı'])
Diab = pd.DataFrame(data=Diab,index=range(768),columns=['Diab'])
age = pd.DataFrame(data=Age,index=range(768),columns=['age'])
OUT = pd.DataFrame(data=out,index=range(768),columns=['out'])
```

```python
t = pd.concat([Preg,Glu],axis=1)
t
```

```python
t1 = pd.concat([t,Blood],axis=1)
t1
```

```python
t2 = pd.concat([t1,Skin],axis=1)
t2
```

```python
t3 = pd.concat([t2,İns],axis=1)
t3
```

```python
t4 = pd.concat([t3,Bmı],axis=1)
t4
```

```python
t5 = pd.concat([t4,Diab],axis=1)
t5
```

```python
t6 = pd.concat([t5,age],axis=1)
t6
```

```python
x_train,x_test,y_train,y_test = train_test_split(t6,OUT,test_size=0.20,random_state=0)
```

```python
y_train
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
```

## LOREG

```python
loreg = LogisticRegression(random_state=0)
loreg.fit(x_train,y_train)
y_pred = loreg.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print(acc)
```

## SVM

```python
svc = SVC(kernel='rbf')
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
acc1 = accuracy_score(y_test,y_pred)
print(acc1)
```

## KNN

```python
knn = KNeighborsClassifier(n_neighbors=15,metric='manhattan')
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
acc2 = accuracy_score(y_test,y_pred)
print(acc2)
```

## Naive Bayes

```python
nb = GaussianNB()
nb.fit(x_train,y_train)
y_pred = nb.predict(x_test)
acc3 = accuracy_score(y_test,y_pred)
print(acc3)
```

## DTREE

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
acc4 = accuracy_score(y_test,y_pred)
print(acc4)
```

## RFC

```python
rfc = RandomForestClassifier(random_state=0)
rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)
acc5 = accuracy_score(y_test,y_pred)
print(acc5)
```
